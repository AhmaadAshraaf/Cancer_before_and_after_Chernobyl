```python
plt.bar(['women','men'],df['gender'].value_counts(),facecolor='b',edgecolor='k')
plt.title('Gender affected men vs. women', size=20)
plt.show()
```

```python
df.value_counts()
```

```python
plt.pie(df['gender'].value_counts(),labels=['Men','Women'],shadow=True,autopct='%.2f%%',colors=['m','r'],explode=[0.1,0])
plt.title('Gender affected men vs. women', size=20)
plt.axis('equal')
plt.show()
```

```python
plt.pie(df['year'].value_counts(),labels=['1985','1986'],shadow=True,autopct='%.2f%%',colors=['m','r'],explode=[0.1,0])
plt.title('Most common year', size=20)
plt.axis('equal')
plt.show()
```

```python
plt.pie(df['cancer_type'].value_counts(),labels=['colon','skin','rectum','thyroid','kidney','lungs','pancreas'],shadow=True,autopct='%.2f%%',colors=['m','r','b','y','g'])
plt.title('% of cancer types', size=20)
plt.axis('equal')
plt.show()
```

```python
X=df[df['year']==1985]['cases'].sum()
X_1=df[df['year']==1986]['cases'].sum()

z=[X,X_1]
ex=[0.2,0.]
Lab=[1985,1986]
col=['#F9FF33','#FF3358']
plt.figure(figsize=(25,8))
plt.title('The percentage of total cases in two years 1985-1986',fontsize=20)
plt.pie(z,labels=Lab,shadow=True,
        startangle=90,
        explode=ex,
        autopct='%1.0F%%',
        rotatelabels=True,
        colors=col,
        labeldistance=1.2)
plt.show( )
```
